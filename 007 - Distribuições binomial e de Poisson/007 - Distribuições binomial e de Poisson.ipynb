# Variávels Aleatórias Discretas: Binomial

**Temos 05 caixas eletrônicos. A probabilidade de cada caixa estar ocupado é de 0,40. Qual função descreve a probabilidade de utilização dos diferentes caixas?**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time

plt.rcParams.update({'font.size': 14})

# número de tentativas e probabilidade de um resultado "bem sucedido" em uma tentativa individual,
# número de simulações
n, p, simulations = 5, 0.40, 10_000
```

A função de probabilidade que descreve a probabilidade de X assumir diferentes valores entre 0 e 5, no exemplo acima, é dada por:
<img src="binomial formula.png">

```python
# Gerando as simulações usando o comando que implementa a função de probabilidade acima
X = np.random.binomial(n, p, simulations)

# E plotando os resultados
plt.figure(figsize = (8, 5))

x_results = list(range(0, n + 1))
probabilities = [(X == i).sum() / simulations for i in range(0, n + 1)]
plt.bar(x_results, probabilities, color = "royalblue", edgecolor = "navy")

# Colocando legendas
plt.title(f"$\itn$ = {n}, $\itp$ = {p:.2f}", fontsize = 25, loc = "center")
plt.xlabel("$\itX$ = $\itx$", fontsize = 20)
plt.ylabel("Probability", fontsize = 20)

plt.savefig("Figura 01.jpg", dpi = 600, bbox_inches = "tight")

plt.show()
```

```python
print("Imprimindo os resultados...")
for result, probability in zip(x_results, probabilities):
    time.sleep(0.5)
    print(f"P(X = {result}) = {probability:.5f}")
    
```

**Suponha que tenhamos uma urna com 4 bolas brancas e 2 bolas pretas. Qual é a probabilidade de se obter duas bolas pretas retiradas sequencialmente da urna e com reposição?**

```python
# Número de tentativas, probabilidade de uma bola preta e número de simulações
n, p, simulations = 2, 2 / 6, 10_000

# Gerando as simulações usando o comando que implementa a função de probabilidade acima
X = np.random.binomial(n, p, simulations)

# E plotando os resultados
plt.figure(figsize = (8, 5))

x_results = list(range(0, n + 1))
probabilities = [(X == i).sum() / simulations for i in range(0, n + 1)]
plt.bar(x_results, probabilities, color = "darkmagenta", edgecolor = "indigo")

# valores do eixo x
plt.xticks([0, 1, 2])

# Colocando legendas
plt.title(f"$\itn$ = {n}, $\itp$ = {p:.2f}", fontsize = 25, loc = "center")
plt.xlabel("$\itX$ = $\itx$", fontsize = 20)
plt.ylabel("Probability", fontsize = 20)

plt.savefig("Figura 02.jpg", dpi = 600, bbox_inches = "tight")


plt.show()
```

```python
print("Imprimindo os resultados...")
for result, probability in zip(x_results, probabilities):
    time.sleep(0.5)
    print(f"P(X = {result}) = {probability:.5f}")
```

**Suponha que um desocupado jogue uma moeda honesta (p = 0,50) para cima 30 vezes e anote seus resultados. Suponha que ele repita este procedimento 10_000 vezes. Gere a distribuição que descreve a probabilidade de diferentes resultados para esses experimentos.**

```python
# Número de tentativas, probabilidade de dar cara e número de simulações
n, p, simulations = 30, 0.50, 10_000

# Gerando as simulações usando o comando que implementa a função de probabilidade acima
X = np.random.binomial(n, p, simulations)

# E plotando os resultados
plt.figure(figsize = (8, 5))


x_results = list(range(0, n + 1))
probabilities = [(X == i).sum() / simulations for i in range(0, n + 1)]
plt.bar(x_results, probabilities, color = "coral", edgecolor = "orangered")

# valores do eixo x
plt.xticks([i for i in range(0, 35, 5)])

# Colocando legendas
plt.title(f"$\itn$ = {n}, $\itp$ = {p:.2f}", fontsize = 25, loc = "center")
plt.xlabel("$\itX$ = $\itx$", fontsize = 20)
plt.ylabel("Probability", fontsize = 20)

plt.savefig("Figura 03.jpg", dpi = 600, bbox_inches = "tight")


plt.show()
```

```python
print("Imprimindo os resultados...")
for result, probability in zip(x_results, probabilities):
    time.sleep(0.5)
    print(f"P(X = {result}) = {probability:.10f}")
```

# Variávels Aleatórias Discretas: Poisson

**O Corinthians tem uma média de 1,3 gols por partida. O técnico acredita que o time tem grandes chances de ganhar o campeonato se marcar mais de 2 gols em pelo menos 4 das 6 primeiras partidas do campeonato. Qual função descreve a probabilidade de marcar diferentes números de gols nessas seis partidas?**

A função de probabilidade que descreve a probabilidade de X assumir diferentes valores entre 0 e 5, no exemplo acima, é dada por:
<img src="poisson formula.png">

```python
# Taxa média de sucessos e número de simulações
par_lambda, simulations = 1.3, 10_000
k = 5

# Gerando as simulações usando o comando que implementa a função de probabilidade acima
W = np.random.poisson(lam = par_lambda, size = simulations)

# E plotando os resultados
plt.figure(figsize = (8, 5))

W_results = list(range(0, k + 1))
probabilities = [(W == i).sum() / simulations for i in range(0, k + 1)]
plt.bar(W_results, probabilities, color = "lightskyblue", edgecolor = "dodgerblue")

# valores do eixo x
plt.xticks([i for i in range(0, k + 1)])

# Colocando legendas
plt.title(f"$\lambda$ = {par_lambda}, $\itN$ simulations = {simulations}", fontsize = 25, loc = "center")
plt.xlabel("$\itX$ = $\itk$", fontsize = 20)
plt.ylabel("Probability", fontsize = 20)


plt.savefig("Figura 04.jpg", dpi = 600, bbox_inches = "tight")


plt.show()
```
