# Variáveis Aleatórias Contínuas

**Gere uma distribuição de QI de uma população (*M* = 100, *DP* = 15).**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Criando a distribuição de QI de uma população hipotética de 1 milhão de habitantes
mu, sigma = 100, 15
QI = np.random.normal(mu, sigma, 1_000_000)

# Plotando minha população hipotética
plt.figure(figsize = (8, 5))

n, bins, patches = plt.hist(QI, sigma, density = True, color = "limegreen", edgecolor = "darkgreen")

# Colocando legendas
plt.title(f"IQ Distribution", fontsize = 25, loc = "center")
plt.xlabel("IQ", fontsize = 20)
plt.ylabel("Density", fontsize = 20)
plt.text(50, .025, r'$\mu = 100,\ \sigma = 15$', fontsize = 20)
plt.xlim(40, 160)
plt.ylim(0, 0.03)
plt.grid(True)

plt.savefig("Figura 07.jpg", dpi = 600, bbox_inches = "tight")


plt.show()
```

```python
# Amostrando 50 participantes de nossa população hipotética e mensurando seus QIs

amostra_N100 = np.random.choice(QI, 100, replace = False)

media, dp = np.mean(amostra_N100), np.std(amostra_N100)
print(f"Média: {media:.5f}, DP: {dp:.5f}")
```

```python
# Plotando minha distribuição amostral
plt.figure(figsize = (8, 5))

n, bins, patches = plt.hist(amostra_N100, density = True, color = "firebrick", edgecolor = "maroon")

# Colocando legendas
plt.title(f"IQ Sample Distribution", fontsize = 25, loc = "center")
plt.xlabel("IQ", fontsize = 20)
plt.ylabel("Density", fontsize = 20)
plt.text(50, .040, f"$\itM$ = {media:.2f}, $\itDP$= {dp:.2f}", fontsize = 20)
plt.xlim(40, 160)
plt.ylim(0, 0.05)
plt.grid(True)

plt.savefig("Figura 08.jpg", dpi = 600, bbox_inches = "tight")


plt.show()
```

**As lâmpadas duram em média 100 horas. Modelar essa variável com uma distribuição exponencial.**

```python
# Parâmetros da simulação
media, n_sims = 100, 10_000

lampadas = np.random.exponential(scale = media, size = 1_000_000)


# Plotando minha distribuição amostral
plt.figure(figsize = (8, 5))

n, bins, patches = plt.hist(lampadas, density = True, color = "darkorchid", edgecolor = "darkviolet")

# Colocando legendas
plt.title(f"Exponential Distribution—Example", fontsize = 25, loc = "center")
plt.xlabel("Lamp Life (Days)", fontsize = 20)
plt.ylabel("Density", fontsize = 20)
plt.xlim(0, 1_000)
plt.ylim(0, 0.006)
plt.grid(True)

plt.savefig("Figura 09.jpg", dpi = 600, bbox_inches = "tight")


plt.show()
```

**Um participante leva em média 450 ms para emitir uma resposta em uma tarefa de decisão lexical. Modelar essa variável com uma distribuição exponencial.**

```python
# Parâmetros da simulação
media, n_sims = 450, 10_000

RTs = np.random.exponential(scale = media, size = 1_000_000)


# Plotando minha distribuição amostral
plt.figure(figsize = (8, 5))

n, bins, patches = plt.hist(RTs, density = True, color = "magenta", edgecolor = "black")

# Colocando legendas
plt.title(f"Exponential Distribution—Example 2", fontsize = 25, loc = "center")
plt.xlabel("Reaction Time (ms)", fontsize = 20)
plt.ylabel("Density", fontsize = 20)
plt.xlim(0, 4_000)
plt.ylim(0, 0.00125)
plt.grid(True)

plt.savefig("Figura 10.jpg", dpi = 600, bbox_inches = "tight")


plt.show()
```
