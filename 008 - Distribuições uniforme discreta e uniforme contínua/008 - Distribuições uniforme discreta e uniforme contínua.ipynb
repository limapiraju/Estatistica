# Distribuição Uniforme Discreta

**Em uma variável aleatória uniforme, os valores possíveis de serem sorteados geralmente ficam confinados num espaço amostral finito [a, b]. Pode ser modelada por meio de uma distribuição uniforme. Exemplo: resultados de um arremesso de um dado honesto.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time

plt.rcParams.update({'font.size': 14})
```

```python
# Valores mínimo e máximo de um dado; número de simulações
a, b = 1, 6
simulacoes = 100_000

# Gerando os dados
dados = np.random.randint(low = a, high = b + 1, size = simulacoes)

# Contando os valores
print("Contando a frequência dos valores...")
for valor in range(a, b + 1):
    time.sleep(0.5)
    print(f"Face '{valor}: {np.count_nonzero(dados == valor, axis = 0)} ocorrências ({np.count_nonzero(dados == valor, axis = 0) / simulacoes})")
```

```python
# Eixos x e y
eixo_x = list(range(a, b + 1))
eixo_y = [(dados == i).sum() / simulacoes for i in range(a, b + 1)]


# Plotando os resultados
plt.figure(figsize = (8, 5))

plt.bar(eixo_x, eixo_y, color = "silver", edgecolor = "dimgray")

# Colocando legendas
plt.title(f"$\itn$ = 1, $\itp$ = 0.1667,\n{simulacoes} simulations", fontsize = 25, loc = "center")
plt.xlabel("Result", fontsize = 20)
plt.ylabel("Probability", fontsize = 20)

plt.savefig("Figura 05.jpg", dpi = 600, bbox_inches = "tight")


plt.show()
```

# Distribuição Uniforme Contínua

**Em uma variável aleatória uniforme, os valores possíveis de serem sorteados geralmente ficam confinados num espaço amostral finito [a, b]. Pode ser modelada por meio de uma distribuição uniforme. Exemplo: valores p sob H0 verdadeira.**

```python
# Valores mínimo e máximo do valor p; número de simulações
minimo, maximo = 0, 1
n_sims = 100_000

# Gerando os dados
valores_p = np.random.uniform(low = minimo, high = maximo, size = n_sims)

np.mean(valores_p), np.std(valores_p)
```

```python
# Plotando os resultados
plt.figure(figsize = (8, 5))

plt.hist(valores_p, density = True, color = "hotpink", edgecolor = "crimson")

# Colocando legendas
plt.title(f"$\itp$($\itp$-value|$H_{0}$),{simulacoes} simulations", fontsize = 25, loc = "center")
plt.xlabel(f"$\itp$-value", fontsize = 20)
plt.ylabel("Density", fontsize = 20)

plt.savefig("Figura 06.jpg", dpi = 600, bbox_inches = "tight")


plt.show()
```
