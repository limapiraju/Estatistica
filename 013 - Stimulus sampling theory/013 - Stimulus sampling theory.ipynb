O objetivo deste *notebook* é descrever brevemente a *stimulus sampling theory*, de William K. Estes (e.g., Estes, 1955a, 1955b; ver também Crowder, 1976). 

Na teoria de Estes, estímulo é entendido não como algo unitário, mas sim como uma população de elementos (e.g., aspectos do contexto), todos com igual importância. Esses estímulos se dividem em dois conjuntos, um (*S*) que pode ser "amostrado" pelo organismo em um dado momento e outro (*S'*) indisponível ao organismo em um dado momento. Os elementos individuais transitam de maneira aleatória entre os dois conjuntos ao longo do tempo (daí o termo *flutuação*; a teoria também é referida como *teoria da flutuação de estímulos* por alguns autores; e.g., Crowder, 1976). 

A curtos intervalos, deve haver pouca troca entre estímulos nos dois conjuntos (i.e., pouca mudança contextual, o que guarda similaridade com teorias modernas de contexto episódico em memória; e.g., Howard & Kahana, 2002). Em intervalos longos, contudo, há uma troca completa de estímulos, de tal modo que um estímulo estar presente no conjunto *S* no tempo *t*(0) pode ser preditivo da presença desse estímulo no tempo *t*(1), mas não no tempo *t*(1.000). 

No contexto de condicionamento, os estímulos podem ser classificados em duas dimensões, a saber, disponíveis para serem condicionados (sim ou não) e amostrados ou não no momento da resposta (e, portanto, condicionados ou não condicionados em um dado momento). Elementos ficam disponíveis em um dado momento com uma probabilidade *p* e ficam indisponíveis com uma probabilidade *q* (*p'*, na notação original de Estes, 1955a). Além disso, elementos escapam do conjunto *S* para o conjunto *S'* com uma probabilidade *j* e escapam do conjunto *S'* para o conjunto *S* com uma probabilidade *j'*. 

Com base nessas informações, é possível derivar as predições do modelo quantitativo. O desempenho do organismo em um teste, segundo a teoria, é diretamente proporcional à fração de elementos condicionados no conjunto disponível (*j*) em relação a todos os elementos daquele conjunto (*S*).

A explicação acima é baseada no resumo verbal de Crowder (1976). A seguir, as simulações se baseiam em Estes (1955a).

```python
# importando bibliotecas importantes
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Estes (1955a), Equação 5
def equation5(p = 0.5, q = 0.5, j = 0.5, j_prime = 0.5):
    """
    Retorna vetor a ser plotado na predição da stimulus sampling theory.
    """
    # quantidades computadas para derivação das predições da teoria
    J = (j_prime / (j + j_prime))
    a = (1 - j - j_prime)
    
    vetor = p * (J - (J - 1) * a ** t) + q * (1 - a ** t) * (1 - J)
    return vetor
```

```python
# proporção de estímulos condicionados a uma resposta S após um período experimental
p01, p02, p03 = 1, .67, .33

# proporção de estímulos condicionados a uma resposta S' após um período experimental
q01, q02, q03 = 0, 0, 0

# probabilidade de que um elemento contido em S escape para S' após um dado período t
j = 0.6

# probabilidade de que um elemento contido em S' escape para S após um dado período t
j_prime = 0.2

# tempo
t = np.linspace(0, 5, 100)

# plotando os gráficos
plt.plot(t, equation5(p01, q01, j, j_prime))
plt.plot(t, equation5(p02, q02, j, j_prime))
plt.plot(t, equation5(p03, q03, j, j_prime))

plt.ylim(-0.1, 1.1)
plt.xlabel(r"$t$", fontsize = 16)
plt.ylabel(r"$p(t)$", fontsize = 16)
plt.title("Curvas de Regressão Espontânea", fontsize = 18)

plt.savefig('013.1 - Estes (1955a), Figura 2, painel superior, curvas de regressão espontânea.jpg', dpi = 600, bbox_inches = "tight")

plt.show()
```

No gráfico acima, a regressão espontânea é similar para os três cenários, pois a proporção de estímulos condicionados a uma resposta *S'* após um período experimental é fixa nos três cenários. No entanto, quanto menor é o valor de *p* em *t*(0), pior é o desempenho do organismo logo após uma sessão de condicionamento, sendo que esse desempenho persistirá sendo o pior ao longo do tempo.

```python
# proporção de estímulos condicionados a uma resposta S após um período experimental
p01, p02, p03 = 1, 1, 1

# proporção de estímulos condicionados a uma resposta S' após um período experimental
q01, q02, q03 = .67, .33, 0

plt.plot(t, equation5(p01, q01, j, j_prime))
plt.plot(t, equation5(p02, q02, j, j_prime))
plt.plot(t, equation5(p03, q03, j, j_prime))

plt.ylim(-0.1, 1.1)
plt.xlabel(r"$t$", fontsize = 16)
plt.ylabel(r"$p(t)$", fontsize = 16)
plt.title("Curvas de Regressão Espontânea", fontsize = 18)

plt.savefig('013.2 - Estes (1955a), Figura 2, painel inferior, curvas de regressão espontânea.jpg', dpi = 600, bbox_inches = "tight")

plt.show()
```

No gráfico acima, a regressão espontânea tem um mesmo ponto de partida para os três cenários, pois a proporção de estímulos condicionados a uma resposta *S* após um período experimental é fixa nos três cenários. No entanto, quanto menor é o valor de *q* em *t*(0), pior é o desempenho assintótico do organismo, uma vez que há o potencial de maior *flutuação* de estímulos entre conjuntos *S* e *S'*.

```python
# proporção de estímulos condicionados a uma resposta S após um período experimental
p01, p02, p03 = .67, .33, 0

# proporção de estímulos condicionados a uma resposta S' após um período experimental
q01, q02, q03 = 1, 1, 1

plt.plot(t, equation5(p01, q01, j, j_prime))
plt.plot(t, equation5(p02, q02, j, j_prime))
plt.plot(t, equation5(p03, q03, j, j_prime))

plt.ylim(-0.1, 1.1)
plt.xlabel(r"$t$", fontsize = 16)
plt.ylabel(r"$p(t)$", fontsize = 16)
plt.title("Curvas de Recuperação Espontânea", fontsize = 18)

plt.savefig('013.3 - Estes (1955a), Figura 3, painel superior, curvas de recuperação espontânea.jpg', dpi = 600, bbox_inches = "tight")

plt.show()
```

No gráfico acima, a recuperação espontânea é similar para os três cenários, pois a proporção de estímulos condicionados a uma resposta *S'* após um período de extinção é fixa nos três cenários. No entanto, quanto maior é o valor de *p* em *t*(0), melhor é o desempenho do organismo logo após uma sessão de extinção, sendo que esse desempenho persistirá sendo o melhor ao longo do tempo.

```python
# proporção de estímulos condicionados a uma resposta S após um período experimental
p01, p02, p03 = 0, 0, 0

# proporção de estímulos condicionados a uma resposta S' após um período experimental
q01, q02, q03 = 1, .67, .33

plt.plot(t, equation5(p01, q01, j, j_prime))
plt.plot(t, equation5(p02, q02, j, j_prime))
plt.plot(t, equation5(p03, q03, j, j_prime))

plt.ylim(-0.1, 1.1)
plt.xlabel(r"$t$", fontsize = 16)
plt.ylabel(r"$p(t)$", fontsize = 16)
plt.title("Curvas de Recuperação Espontânea", fontsize = 18)

plt.savefig('013.4 - Estes (1955a), Figura 3, painel inferior, curvas de recuperação espontânea.jpg', dpi = 600, bbox_inches = "tight")

plt.show()
```

No gráfico acima, a recuperação espontânea tem início em 0 para os três cenários, pois não há mais estímulos condicionados a uma resposta alvo em *S*. No entanto, quanto maior é o valor de *q* em *t*(0), maior será o grau de recuperação espontânea do organismo com o tempo, pois há mais estímulos passíveis de *flutuarem* de *S'* para *S*.
