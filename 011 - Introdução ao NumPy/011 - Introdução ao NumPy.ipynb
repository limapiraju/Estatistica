**Reshape, concatenação e divisão de arrays**

```python
grid = np.arange(1, 10)
print(grid)

grid = grid.reshape((3, 3))
print(grid)
```

```python
x = np.array([1, 2, 3])

# vetor linha via reshape
print(x.reshape((1, 3)))

# vetor coluna via reshape
print(x.reshape((3, 1)))
```

```python
# concatenção de dois arrays
x = np.array([1, 2, 3])
y = np.array([3, 2, 1])
np.concatenate([x, y])
```

```python
z = [99, 99, 99]
print(np.concatenate([x, y, z]))
```

```python
grid = np.array([[1, 2, 3], [4, 5, 6]])

# concatenar a partir do primeiro eixo (axis = 0)
np.concatenate([grid, grid])
```

```python
# concatenar a partir do segundo eixo (axis = 1)
np.concatenate([grid, grid], axis = 1)
```

```python
# divisão de arrays
x = [1, 2, 3, 99, 99, 3, 2, 1]
x1, x2, x3 = np.split(x, [3, 5]) # divide no índice 3 e depois no 5
print(x1, x2, x3)
```

```python
grid = np.arange(16).reshape((4, 4))
print(grid)

upper, lower = np.vsplit(grid, [2]) # divide ao meio (row-wise)
print(upper)
print(lower)
```

```python
left, right = np.hsplit(grid, [2]) # divide ao meio (column-wise)
print(left)
print(right)
```

**Matemática**

```python
# operações aritméticas com escalares
x = np.arange(4)
print("x =", x)
print("x + 5 =", x + 5)
print("x - 5 =", x - 5)
print("x * 2 =", x * 2)
print("x / 2 =", x / 2)
print("x // 2 =", x // 2) # floor division
print("-x = ", -x)
print("x ** 2 = ", x ** 2)
print("x % 2 = ", x % 2)
```

```python
# operações entre vetores ou matrizes
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])

matriz1 = np.random.randint(1, 10, size = (3, 3))
matriz2 = np.random.randint(1, 10, size = (3, 3))

print(matriz1)
print(matriz2)
```

```python
# adição de vetores
print(array1 + array2)
print(np.add(array1, array2))
```

```python
# subtração de vetores
print(array1 - array2)
print(np.subtract(array1, array2))
```

```python
# multiplicação de vetor por escalar
print(array1 * 3)
print(np.multiply(array1, 3))
```

```python
# multiplicação de vetores (element-wise)
print(array1 * array2)
print(np.multiply(array1, array2))
```

```python
# multiplicação (produto vetorial) = soma dos produtos vetoriais
print(np.sum(array1 * array2))
print(np.dot(array1, array2))
```

```python
# multiplicação cruzada (cross-product)
print(np.cross(array1, array2))
```

```python
# multiplicação matricial
print(matriz1 @ matriz2)
print(np.matmul(matriz1, matriz2))
```

```python
# divisão (element-wise)
print(array1 / array2)
print(np.divide(array1, array2))
```

```python
# módulo de um vetor (raiz quadrada da soma dos produtos do vetor por ele mesmo)
print(sum(array1 ** 2) ** 0.5)
print(np.linalg.norm(array1))
```

```python
# determinante de uma matriz
print(np.linalg.det(matriz1))
print(np.linalg.det(matriz2))
```

```python
# matriz inversa
print(np.linalg.inv(matriz1))
print(np.linalg.inv(matriz2))
```

```python
# multiplicando uma matriz pela sua inversa resulta em uma matriz identidade
print(np.matmul(matriz1, np.linalg.inv(matriz1)).round(3))
```

```python
# transposta de uma matriz
print(matriz1.T)
print(np.transpose(matriz1))
```

```python
# Funções trigonométricas
# Trigonometric functions
theta = np.linspace(0, np.pi, 3)
print("theta = ", theta)
print("sin(theta) = ", np.sin(theta))
print("cos(theta) = ", np.cos(theta))
print("tan(theta) = ", np.tan(theta))

x = [-1, 0, 1]
print("x = ", x)
print("arcsin(x) = ", np.arcsin(x))
print("arccos(x) = ", np.arccos(x))
print("arctan(x) = ", np.arctan(x))
```

```python
# Expoentes e logaritmos

x = [1, 2, 3]
print("x =", x)
print("e^x =", np.exp(x))
print("2^x =", np.exp2(x))
print("3^x =", np.power(3, x))

x = [1, 2, 4, 10]
print("x =", x)
print("ln(x) =", np.log(x))
print("log2(x) =", np.log2(x))
print("log10(x) =", np.log10(x))
```

```python
# Outer product
x = np.arange(1, 11)
np.multiply.outer(x, x)
```

```python
# álgebra linear
a = np.ones((2,3))
print(a)

b = np.full((3,2), 2)
print(b)

# produto de matrizes
np.matmul(a,b)
```

**Filtros**

```python
np.random.seed(42)

array = np.random.randint(0, 100, size = 10)
array
```

```python
array > 10
```

```python
# comparações múltiplas (precisam estar entre parênteses)
# & = and
# | = or
# ~ = not
(array > 50) & (array < 75)
```

```python
# aplicando um filtro
array[(array > 50) & (array < 75)]
```

```python
# np.where() = retorna os índices em que uma condição for verdadeira
np.where((array > 50) & (array < 75))
```

```python
# modificando os valores do array com base no True ou False do np.where
np.where((array > 50) & (array < 75), "Papai Noel é Top", "Papai Noel é Gordo")
```

```python
np.all(array.reshape(2, 5) < 75, axis = 1)
```

```python
np.all(array.reshape(2, 5) < 75, axis = 0)
```

```python
np.all(array.reshape(2, 5) < 75)
```

**Manipulação de arrays**

```python
# concatenate
a = np.array([[1, 2, 3], [1, 2, 3]])
b = np.array([[4, 5, 6], [4, 5, 6]])

c = np.concatenate([a, b], axis = 0) # eixo das linhas; empilha arrays
d = np.concatenate([a, b], axis = 1) # eixo das colunas; estende arrays

print(c)
print(d)
```

```python
# vstack = empilha arrays verticalmente, mesmo que eles tenham apenas uma única direção
e = np.array([1, 2, 3])
f = np.array([4, 5, 6])

g = np.vstack([e, f])
g
```

```python
# hstack = estende arrays horizontalmente
h = np.hstack([e, f])
h
```
